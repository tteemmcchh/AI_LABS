# Подготовка данных дла обучения

```python
import pandas as pd
import numpy as np

titanic_data = pd.read_csv("train.csv")

# Проверяем наличие пропущенных значений в каждом столбце
titanic_data.isnull().sum()

# Удаляем столбец "Cabin" так как он нам не нужен для обучения модели
titanic_data = titanic_data.drop(columns='Cabin', axis=1)

# Заполняем пропущенные значения столбуца "Embarked" модой по выборке
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)

#Заполняем пропущенные значения средним значсчением по выборке
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)

# Меняем значения в столбце "Sex" на нули и единицы, в столбце "Embarked" на 0.1.2
titanic_data.replace({'Sex':{'male':0.0,'female':1.0}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

# Процесс обучения и тестирования

Для тестовой и валидационной выборки нас  не интересуют некоторы поля приведенные в исходных данных, поэтому отбросим данные, оп нашему мнению, не влияющие на выживаемость.

```python
from keras.models import Sequential
from keras.layers import Dense,BatchNormalization,Dropout

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X = titanic_data.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = titanic_data['Survived']
```

```python
# Разбиваем текущию выборку на тестовую и тренировочную
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
# Нормализация данных
scaler = StandardScaler() # Стандартизация
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(1, activation='relu'))

model.compile(loss='mean_squared_error', metrics=['accuracy'])
history = model.fit(X_train, Y_train, epochs=10, validation_split=0.2)
```

```python
print(history.history.keys())

# Графики обучения
plt.plot(history.history['loss'], label='Обучение')
plt.plot(history.history['val_loss'], label='Валидация')
plt.title('Обучение на задаче классификации \n используя Dropout()')
plt.xlabel('Epochs (Эпохи)')
plt.yticks(np.arange(min(Y_test), max(Y_test)-0.5, 0.05))
plt.ylabel('Loss (Функция потерь)')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='Обучение')
plt.plot(history.history['val_accuracy'], label='Валидация')
plt.title('Обучение на задаче классификации')
plt.ylabel('Accuracy (точность)')
plt.xlabel('Epoch (Эпохи)')
plt.legend()
#plt.legend(['train', 'validation'], loc='lower right')
plt.show()
```

```python
# Оценка модели на тестовых данных
test_loss = model.evaluate(X_test, Y_test)
print(f"Метрика средней квадратичной ошибки (MSE): {test_loss[0]}")
```
